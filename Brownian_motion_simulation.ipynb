```python
import numpy as np
import matplotlib.pyplot as plt

T = 1.0
N = 252
dt = T / N
mu = 0.05
sigma = 0.2
S0 = 100

np.random.seed(42)
Z = np.random.normal(0, 1, N)

S = np.zeros(N)
S[0] = S0

for t in range(1, N):
    S[t] = S[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z[t])

plt.figure(figsize=(10,5))
plt.plot(S)
plt.title("Simulation d'un mouvement brownien gomtrique")
plt.xlabel("Temps (jours)")
plt.ylabel("Prix de l'actif")
plt.show()
```
